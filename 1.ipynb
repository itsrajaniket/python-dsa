### Write a Python program to do arithmetical operations add

```python
num1 =int(input("Enter first number  :"))
num2 = int(input("Enter second number : "))

sum = num1 + num2 

print(f"Sum of {num1} and {num2} = {sum} ")
```

### Write a Python program to do arithmetical operations division.

```python
num1 = int(input("enter first number :"))
num2 = int(input("enter second number :"))
if num2 == 0 :
    print ("number cant be 0")
else :
    result = num1 / num2
    print(f"division of {num1} and {num2} = {result} ")
```

### Write a Python program to find the area of a triangle

```python
base = int(input("Enter the base of the triangle: "))
height = int(input("Enter the height of the triangle: "))

area = (base * height) / 2
print(f"Area of triangle = {area}")
```

### Write a Python program to swap two variables

```python
num1 =int(input("Enter first number - A:"))
num2 = int(input("Enter second number - B: "))
print(f"Original values: A = {num1}, B = {num2}")
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2

print(f"Swapped values: A={num1} and B={num2}")
```
