### 1 Write a Python program to do arithmetical operations add

```python
num1 =int(input("Enter first number  :"))
num2 = int(input("Enter second number : "))

sum = num1 + num2 

print(f"Sum of {num1} and {num2} = {sum} ")
```

### 2 Write a Python program to do arithmetical operations division.

```python
num1 = int(input("enter first number :"))
num2 = int(input("enter second number :"))
if num2 == 0 :
    print ("number cant be 0")
else :
    result = num1 / num2
    print(f"division of {num1} and {num2} = {result} ")
```

### 3 Write a Python program to find the area of a triangle

```python
base = int(input("Enter the base of the triangle: "))
height = int(input("Enter the height of the triangle: "))

area = (base * height) / 2
print(f"Area of triangle = {area}")
```

### 4 Write a Python program to swap two variables

```python
num1 =int(input("Enter first number - A:"))
num2 = int(input("Enter second number - B: "))
print(f"Original values: A = {num1}, B = {num2}")
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2

print(f"Swapped values: A={num1} and B={num2}")
```

### 6 Write a Python program to convert kilometers to miles.

```python
km = int(input("enter the km" ))
#  Conversion factor: 1 kilometer = 0.621371 miles
m = km * 0.621371
print(f"{km} kilometers will be = {m: .2f} miles")
```

#### Write a Python program to convert Celsius to Fahrenheit.
- Conversion formula: Fahrenheit = (Celsius * 9/5) + 32

```python
c = int(input("Enter the celsius:"))
f = (c*9/5) + 32
print(f"{c} Celsius will be = {f} Fahrenheit")
```
